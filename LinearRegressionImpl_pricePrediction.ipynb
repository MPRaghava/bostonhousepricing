```python
data_filled  = data.fillna(0)
data_filled
```

```python
##Independent and Dependent Features 
#data = data.drop('MEDV',axis =1)
X = data_filled.iloc[:,:-1]
y = data_filled.iloc[:,-1]
```

```python
X.head()
```

```python
#data_cleaned = data.dropna(subset=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','Price'],how='all')
```

```python
##Train test Split

from sklearn.model_selection import train_test_split

X_train,X_test ,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=42)

```

```python
## Linear Regression 

from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(X_train,y_train)
```

```python
#How much is increased or decrease in dependent variables[CRIM, RM..etc] with respect to the Target feature 
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
##Prediction

reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred)
```

```python
residuals = y_test - reg_pred
```

```python
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
```
